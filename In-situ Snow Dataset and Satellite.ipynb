```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

data_dir = Path(".")

# Insitu data
df = pd.read_csv(data_dir / "in_Situ" / "Insitu_Snow_HB_2023_2025.csv",skiprows=2)
df['Date'] = pd.to_datetime(df['Date'])

# in â†’ cm (1 inch = 2.54 cm)
df['snow_depth_cm'] = df['SNWD.I-1 (in) '] * 2.54

df.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['snow_depth_cm'], marker='o')
plt.xlabel("Date")
plt.ylabel("Snow Depth (cm)")
plt.title("Hubbard Brook Snow Depth (2023â€“2025)")
plt.grid(True)
plt.show()
```

My initial plan was to perform satelliteâ€“in situ validation using snow observations collected at the Kingman Farm field site. However, the Kingman Farm In situ measurements are not publicly distributable due to data-sharing restrictions and project policy.   
To address this limitation while maintaining reproducibility, I used an alternative public in-situ snow dataset from Hubbard Brook, NH published by USDA.
